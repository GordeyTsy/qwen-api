```python
import os
import pathlib
import shutil
import subprocess
import sys

print('BOOT: qwen-api bootstrap started', flush=True)
repo_url = os.getenv('QWEN_REPO_URL', 'https://github.com/GordeyTsy/qwen-api.git')
repo_ref = os.getenv('QWEN_REPO_REF', 'main')
repo_dir = pathlib.Path('/kaggle/working/qwen-api')
if repo_dir.exists():
    shutil.rmtree(repo_dir)
subprocess.run(['git', 'clone', '--depth', '1', '--branch', repo_ref, repo_url, str(repo_dir)], check=True)
subprocess.run([sys.executable, '-m', 'pip', 'install', '-e', str(repo_dir)], check=True)
subprocess.run([sys.executable, '-m', 'qwen_api.kaggle_runner'], check=True)
```
